```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy import ndimage
```

```python
pic = plt.imread('019.png')
pic.shape
plt.imshow(pic)
```

```python
pic.shape
```

```python
size_changed = cv2.resize(pic,(263,192))
plt.imshow(size_changed)
```

```python
gray_pic = rgb2gray(size_changed)
plt.imshow(gray_pic, cmap='gray')
```

```python
gray_pic.shape
```

```python
D_Gr_Bu = gray_pic.reshape(gray_pic.shape[0]*gray_pic.shape[1])
for i in range(D_Gr_Bu.shape[0]):
    if D_Gr_Bu[i] > D_Gr_Bu.mean():
        D_Gr_Bu[i] = 1
    else:
        D_Gr_Bu[i] = 0
gray_pic = D_Gr_Bu.reshape(gray_pic.shape[0],gray_pic.shape[1])
print("Differentiate Buildings and Greenery using opencv in satellite images")
plt.imshow(gray_pic, cmap='gray')
```
